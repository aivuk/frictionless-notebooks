# Loading a data package


## Using Ckan Control

```python
from frictionless import portals, Package

control = portals.CkanControl()
package = Package('http://localhost:5000/dataset/validation-test', control=control)
package.resources[0]
```

# Or we could have used the 'baseurl' parameter to control

```python
# In that case we just need to pass the dataset name

control = portals.CkanControl(baseurl='http://localhost:5000')
package = Package('validation-test', control=control)
package
```

# Fetching packages using a Catalog

```python
from frictionless import portals, Package, Resource, Catalog

control = portals.CkanControl(baseurl='https://legado.dados.gov.br', ignore_package_errors=True, num_packages=1000)
catalog = Catalog(control=control)


len(catalog.packages)
```

## Using CKAN search 

 * CKAN search endpoint documentation: https://docs.ckan.org/en/2.9/api/#ckan.logic.action.get.package_search

```python
control = portals.CkanControl(baseurl='https://legado.dados.gov.br',
                              ignore_package_errors=True,
                              num_packages=1000,
                              search={
                                  'q': 'bolsa'
                              })
catalog = Catalog(control=control)
len(catalog.packages)
```

# Getting all packages from an Organization

```python
control = portals.CkanControl(baseurl='https://legado.dados.gov.br',
                              ignore_package_errors=True,
                              group_id='agricultura-extrativismo-e-pesca',
                              num_packages=1000,
                              )
catalog = Catalog(control=control)
catalog.packages
```
